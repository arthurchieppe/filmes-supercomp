```python
import pandas as pd
import random
```

```python
# Generate random pandas dataframe with 1000 rows and columns ['n_movies', 'n_cat', 'time', "n_output"]

df = pd.DataFrame(columns=['n_movies', 'n_cat', 'time', "n_output"])
for i in range(1000):
    df.loc[len(df)] = [random.randint(1, 1000),
                       random.randint(1, 1000),
                       random.randint(1, 1000),
                       random.randint(1, 1000)]

```

```python
df
```

```python
# Sort df first by n_movies then by n_cat

df.sort_values(by=['n_movies', 'n_cat'])

# Order index after sorting

df = df.sort_values(by=['n_movies', 'n_cat']).reset_index(drop=True)
```
